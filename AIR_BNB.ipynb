dividing data-set in 'X' and 'Y'

```python
x=data.iloc[:,:-1] #selecting x feature from dataset
x.values
```

```python
y=data.iloc[:,-1] # selecting y feature from dataset
y.values
```

 Divide data set for train test phase using train_test_split method for ml model

```python
from sklearn.model_selection import train_test_split #importing train_test method
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=10) #divide dataset from training and testing phase
```

Trainnig our model with decision tree

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()                      # create object of model                          # fit the model
```

```python
model=DecisionTreeClassifier() # create object of model
model.fit(x_train,y_train)  # fit the model
ypred= model.predict(x_test)
model.score(x_train, y_train)  # accuracy score of model
```

RESULT >> ACCURACY OF MODEL IS 0.87 % percent

## **Evaluating model and test results**

```python
from sklearn.metrics import confusion_matrix as cm #importing evaluating matrix
r = cm(y_test, ypred)
print(r)
```

```python
from sklearn.metrics import classification_report as cr
r = cr(y_test, ypred)
print(r)
```
